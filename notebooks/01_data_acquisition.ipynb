# Data Acquisition

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

back to root to call functions from helpers.py file

```python
import os
from pathlib import Path
print(Path.cwd())
os.chdir(Path('..').resolve())
from src.utils.helpers import save
print(Path.cwd())
```

load train data

```python
df_train = pd.read_csv(r'data/raw/train.csv', header=None, names=['review_target', 'review_title', 'review_content'], dtype=str, quoting=0)
df_train.head()
```

```python
df_train.info()
```

Due to the large dataset size, we will work on a random 80k for training and 20k for validation subset for faster iteration and development.  


```python
sample_train_valid = df_train.sample(n=100000, random_state=42).reset_index(drop=True)
sample_train_valid.head()
```

```python
sample_train_valid.info()
```

split train dataset to train 80% and valid 20%

```python
sample_train, sample_valid = train_test_split(sample_train_valid, test_size=0.2, random_state=42, stratify=sample_train_valid['review_target'])
print(f"Training set size: {len(sample_train)}")
print(f"Validation set size: {len(sample_valid)}")
```

save train dataset 

```python
save(base_path='data/samples', df=sample_train, df_name='sample_train.csv')
```

```python
save(base_path='data/samples', df=sample_valid, df_name='sample_valid.csv')
```
